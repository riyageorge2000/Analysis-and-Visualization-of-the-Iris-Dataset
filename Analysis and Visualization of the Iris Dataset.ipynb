# Analysis and Visualization of the Iris Dataset

The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository. It includes three iris species with 50 samples each as well as some properties about each flower.

The iris dataset is a widely used dataset. The Iris dataset contains 150 observations of iris flowers, with 50 observations for each of the three species (setosa, versicolor, and virginica). Each observation has four features or variables: sepal length, sepal width, petal length, and petal width. These features are all numeric and continuous. The dataset contains no missing or null values.

Detailed description of the columns:
Sepal length (in cm): The length of the sepal (the outermost part of the flower that protects the inner petals and reproductive organs).
Sepal width (in cm): The width of the sepal.
Petal length (in cm): The length of the petal (the brightly colored part of the flower that attracts pollinators).
Petal width (in cm): The width of the petal.

```python
# Correlation analysis
# Getting correlation matrix

d.corr()
```

The correlations are between different measurements of the iris flower: Sepal Length, Sepal Width, Petal Length, and Petal Width. The values range from -1 to 1, with positive values indicating a positive correlation (as one variable increases, so does the other) and negative values indicating a negative correlation (as one variable increases, the other decreases).

Id and and all other columns are perfectly correlated with itself, as expected, with a correlation coefficient of 1.
Sepal Length and Petal Length have a strong positive correlation of 0.871754, indicating that longer petals are associated with longer sepals. This correlation is also visually apparent in a scatter plot.
Petal Length and Petal Width also have a strong positive correlation of 0.962757, indicating that larger petals tend to be wider. Again, this relationship is visually apparent in a scatter plot.
Sepal Length and Petal Width have a moderate positive correlation of 0.817954, indicating that longer sepals tend to be associated with wider petals.
Sepal Length and Sepal Width have a weak negative correlation of -0.109369, indicating that longer sepals tend to be associated with narrower sepals, but the correlation is not very strong.
Sepal Width and Petal Length have a weak negative correlation of -0.420516, indicating that wider sepals tend to be associated with shorter petals.
Sepal Width and Petal Width have a weak negative correlation of -0.356544, indicating that wider sepals tend to be associated with narrower petals.

```python
# Create a scatterplot of sepal length vs. sepal width

sns.scatterplot(data=d, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
# Set the title and axes labels
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
# Show the plot
plt.show()
```
